```python
import pandas as pd
from openpyxl import load_workbook

def csv_a_excel_con_texto(ruta_csv, ruta_excel):
    """Convierte un CSV a Excel conservando todos los caracteres como texto."""
    
    # 1. Leer CSV forzando todo como texto
    df = pd.read_csv(ruta_csv, dtype=str, encoding="utf-8")
    
    # 2. Guardar en Excel
    with pd.ExcelWriter(ruta_excel, engine="openpyxl") as writer:
        df.to_excel(writer, index=False, sheet_name="Datos")
    
    # 3. Aplicar formato de texto a todas las celdas
    wb = load_workbook(ruta_excel)
    ws = wb.active
    
    for row in ws.iter_rows():
        for cell in row:
            cell.number_format = "@"
    
    wb.save(ruta_excel)

# Ejecutar para los 3 archivos
archivos = {
    "errores_validation.csv": "errores_validation.xlsx",
    "errores_test.csv": "errores_test.xlsx",
    "errores_train.csv": "errores_train.xlsx"
}

for csv, xlsx in archivos.items():
    ruta_csv = f"../data/COMPLETE/V3/{csv}"
    ruta_excel = f"../data/COMPLETE/V3/{xlsx}"
    csv_a_excel_con_texto(ruta_csv, ruta_excel)
    print(f"Convertido: {csv} -> {xlsx}")
```
